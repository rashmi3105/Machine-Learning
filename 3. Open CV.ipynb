```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('c:/Users/User/Documents/Python files/images/test.jpg')
```

```python
print(img.shape)
```

### open cv is going to read image in BGR format
     Which means it is going to consider first channel as BLUE, the second channel as GREEN and the third channel as RED.
     The original image was in RGB format so open CV changed R ---> B , G ---> G and B ---> R
##### Data is always stored in BGR format

```python
plt.style.use('seaborn')
plt.imshow(img)
plt.show
```

##### But we want to display color in RGB format

```python
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.axis("off")    #to remove the numbers on the picture
plt.imshow(rgb_image)
plt.show
```

```python
plt.axis("off")
plt.imshow(gray_image)
plt.show
```

##### To crop a picture

```python
# To check for dimensions
plt.imshow(rgb_image)
plt.show
```

```python
crop_image = rgb_image[150:600,200:600,:]   # In this first attribute is row and second one is column.
plt.axis("off")
plt.imshow(crop_image)
plt.show
```

```python
crop_image.shape
```

### To calculate pixels

```python
total_pixels = crop_image.shape[0]*crop_image.shape[1]
print(total_pixels)
```

### To resize image

```python
resized_image = cv2.resize(crop_image,(300,300))
plt.imshow(resized_image)
plt.title("Profile Photo "+ str(resized_image.shape))
plt.show
```

# Data Visulization

```python
import matplotlib.pyplot as plt
import numpy as np
```

plt is an object that has certain methods

```python
x = np.array([1, 2, 3, 4])
y = x**2 + 3     #x**2 means x square
print(x,y)
```

```python
plt.plot(x,y)
plt.show
```

### To find out styles available for plotting

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x,y)
plt.show
```

### You can also give title to a plot
     label will not work unless plt.legend is used

```python
plt.title("My line graph")
plt.plot(x,y,label="Stock Price")
plt.legend()  
plt.show
```

### If you want another plot

```python
plt.plot(x,x**2,label="Stock Price 1")
plt.plot(x,y,label="Stock Price 2")
plt.legend()  
plt.show
```

### To add x and y label

```python
plt.plot(x,x**2,label="Stock Price 1")
plt.plot(x,y,label="Stock Price 2")
plt.legend()
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.show
```

## Scatter Plot
   Replace plot with scatter

```python
plt.scatter(x,x**2,label="Stock Price 1")
plt.scatter(x,y,label="Stock Price 2")
plt.legend()
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.show
```
